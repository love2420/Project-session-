```python
Content = []
for i in range (1,11):
    URL= f'http://quotes.toscrape.com/page/{i}/'
    Response= requests.get(URL)
    Response= Response.content
    soup= BeautifulSoup(Response,'html.parser')
    contents = soup.find_all('div',class_= 'quote')

for content in contents:
    Quote= content.find('span',class_='text').text.strip()
    print(Quote)
    
for content in contents:
    Name = content.find('small',class_='author').text.strip()
    print(Name)
    
    soup.select('.tag-items')
for tag in soup.select('.tag'):
    print(tag.text)
    
    Content.append([Quote, Name, tag])
    print(Content)
```

```python
df = pd.DataFrame(Content,columns = ['Quotes','Author','Tags'])
```
