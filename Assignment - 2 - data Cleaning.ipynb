```python
df = pd.read_csv("data cleaning.csv")
df['Name'] = df['Name'].fillna(df['Email'].str.extract('(\w+)[\.\w]*@', expand=False).str.capitalize())
age_mean = df['Age'].mean()
df['Age'].fillna(age_mean, inplace=True)
df['Phone'] = df.Phone.str.strip('(')
df['Phone'] = df['Phone'].str[-10:]
df['Phone'] = '91-'+df['Phone']  
df.dropna(subset=['Name', 'Email'], inplace=True)
Top_10 = df.nlargest(10, 'Salary').style.background_gradient()
df['Email'] = df['Email'].fillna('support@dataisgood.com')
corrections = {'SLES': 'Sales',
                    'SALES': 'Sales',
                    'DaTa SciENce': 'Data Science',
                    'hR': 'Hr',
                    'HR':'Hr',
                     'IT DeVlopemENT' : 'IT Development',
                     'OperarYION':'operation'}
df['Department'] = df['Department'].replace(corrections)
df['Department'] = df['Department'].str.title()
df['Address'] = df['Address'].fillna('Address not available')
```
